- Bereken het gemiddelde. De mogelijke waarden bij iedere score is 0 of 1, maar de gemiddelden zijn hier decimale (gebroken) getallen. 
  * Hoe dien je dit te interpreteren?
  * Als het gemiddelde van een vraag hoog is, kun je dan de volgende *kwalitatieve* uitspraak doen: "De vraag was makkelijk"? Waarom (niet)?
  

_De getallen berekenen wat er gemiddeld gescoord is per vraag. Het gemiddelde kan een decimaal (gebroken) getal zijn, maar kan niet optreden als waarde in de dataset (deze is binair, 0 of 1). Het getal drukt uit welke deel van de studenten de vraag goed hadden beantwoord. Een gemiddelde van 0 betekent dat 0% van de studenten de vraag goed hadden. Een gemiddelde van 0.4 betekent: 40% had de vraag goed._

_Een hoog gemiddelde betekent dat een groot percentage de vraag goed had. Dit maakt het aannemelijk (maar niet per se noodzakelijk) dat het een makkelijke vraag was. Je kunt dit echter niet met zekerheid zeggen, want andere factoren kunnen ervoor zorgen dat het een gemiddelde hoog is, bijv:_
1. _De studenten waren goed, hiertoe dien je te vergelijkene met de andere vragen/klassen/jaren._
2. _Misschien is de vraag voorgezegd tijdens het afnemen. (hiertoe dien je te vergelijkenen met de andere klassen/jaren)._

  1. cijfer = $\frac{9}{30}(\sum_{i=1}^{40}{Score.i}-10) + 1$

- Laten we aannemen dat de cijfers normaal verdeeld zijn. Welke parameters heeft een normaalverdeling?


Een normaalverdeling wordt beschreven door 2 parameters:
1. Het gemiddelde $\mu$
2. De standaard deviatie $\sigma$

Als we de cijferdistributie willen benaderen met een normaalverdeling, dan geldt:
1. $\mu = $ uitslag_csn.Cijfer.mean()
2. $\sigma = $ uitslag_csn.Cijfer.stdev()

- Wat zijn deze parameters voor onze cijfer-verdeling? Sla deze verdeling op in een variable **C** en plot **C**. Gebruik hiervoor `norm` van `SciPy`.

```python
mu_cijfers = uitslag_csn.Cijfer.mean()
sigma_cijfers = uitslag_csn.Cijfer.std()
normal_cijfers = sp.stats.norm(mu_cijfers, sigma_cijfers)

x = np.linspace(-5,15,1000)
y = normal_cijfers.pdf(x)
plt.plot(x,y)
plt.show()
```

- Bereken:
 * $\Pr(C < 6)$. Tip: gebruik `cdf()`
 * $\Pr(C > 8,5)$
 * $\Pr(C > 10)$. Leg uit waarom dit groter dan 0 is.
 * $\Pr(6 < C < 8,5)$.
 * $\Pr(C < 6 \lor C > 8,5) \>$ ($\lor$ is wiskundige notatie voor OR)


Om kansen te berekenen moeten we oppervlakten onder de normaalverdeling berekenen. Dit kunnen we realiseren met de `cdf()` functie. Deze functie retourneert de oppervlakte onder de curve t/m het ingevoerde argument. Dus, $\Pr(C < 6)$ = `C.cdf(6)`.

Volgens de regels van kansreken geldt verder:
- $\Pr(C > 8.5) = 1 - \Pr(C < 8.5)$
- $\Pr(6 < C < 8.5) = \Pr(C < 8.5) - \Pr(C < 6)$
- $\Pr(C < 6 \lor C > 8.5) = \Pr(C < 6) + \Pr(C > 8.5) = \Pr(C < 6) + 1 - \Pr(C < 8.5)$

De laatste regel geldt, omdat de gebeurtenissen $C < 6$ en $C > 8.5)$ disjunct zijn.

$\Pr(C > 10)$ zal groter zijn dan 0, omdat de "staarten" van de normaalverdeling oneindig doorgaan. De kansen gaan wel heel snel naar 0, maar zullen nooit precies 0 zijn.

```python
c_below_6 = normal_cijfers.cdf(6)
c_below_8 = normal_cijfers.cdf(8.5)
c_above_8 = 1 - c_below_8
c_between_6_and_8 = c_below_8 - c_below_6
c_outside_6_and_8 = 1 - c_between_6_and_8     # of c_below_6 + c_above_8
c_above_10 = 1 - normal_cijfers.cdf(10)

pd.Series(
    [c_below_6, c_above_8, c_above_10, c_between_6_and_8, c_outside_6_and_8],
    ["P(C < 6)", "P(C > 8.5)", "P(C > 10)", "P(6 < C < 8.5)", "P(C < 6 OR C > 8.5)"]
)
```

- Plot nogmaals de normaalverdeling, dit maal met het oppervlak 6 < cijfer < 8,5 ingekleurd. Tip: gebruik `fill_between()`)


```python
plt.plot(x,y)
plt.fill_between(x,y, where=sp.logical_and(x>6, x<8.5))
plt.show()
```

- Een student verkondigt trots dat zij een 8,2 heeft gehaald voor PROG. Ze is benieuwd hoeveel procent van de studenten het beter heeft gemaakt dan zij.
  * Bepaal de z-score van deze student op basis van de gehele populatie. Wat zegt deze score?
  * Bepaal per klas hoeveel procent van de klas beter heeft gescoord. Wat zegt deze score?

```python
from scipy import stats
```

```python
vEindCijferZ = vEindCijfer.dropna('index', subset=['PROG'])
vEindCijferZ.head()
```

```python
vEindCijferZ['z-score'] = stats.zscore(vEindCijferZ[['PROG']])
vEindCijferZ
```

```python
vEindCijferZ[vEindCijferZ['PROG'] == 8.2][['z-score']].head()
```

```python
vKlasStat = vEindCijferZ[['z-score']].groupby(['Klas']).count()
vKlasStat.rename(columns={'z-score':'total'}, inplace=True)
vKlasStat
```

```python
vKlasStat['boven8.2'] = vEindCijferZ[vEindCijferZ['z-score'] > 1.328938]['z-score'].groupby(level='Klas').count()
vKlasStat.fillna(0, inplace=True)
vKlasStat['rel'] = vKlasStat['boven8.2']/vKlasStat['total'] * 100
vKlasStat
```
